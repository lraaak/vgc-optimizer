```python
import sys
import pandas as pd
sys.path.append('../src')
from battle_engine import BattlePokemon, BattleState
```

```python
# 1. Load our clustered data to act as the base stats
df_pokemon = pd.read_csv('../data/processed/clustered_pokemon_data.csv').set_index('name')
incineroar_row = df_pokemon.loc['incineroar']

# 2. Instantiate a State Object (ignoring moves for this quick test)
incineroar = BattlePokemon(
    name='incineroar', 
    pokemon_row=incineroar_row, 
    moves=[], 
    ability='Intimidate'
)

# 3. Test the State Machine
print(f"--- {incineroar.name.upper()} STATE TEST ---")
print(f"Max HP (Level 50): {incineroar.max_hp}")
print(f"Base Attack (Level 50): {incineroar.get_stat('attack')}")

print("\n[Action] Opponent uses Intimidate!")
incineroar.change_stage('attack', -1)
print(f"Modified Attack (-1 Stage): {incineroar.get_stat('attack')}")

print("\n[Action] Incineroar gets Burned!")
incineroar.status = 'BRN'
print(f"Modified Attack (-1 Stage + Burn): {incineroar.get_stat('attack')}")

print("\n[Action] Incineroar takes 100 damage!")
incineroar.apply_damage(100)
print(f"Current HP: {incineroar.current_hp} / {incineroar.max_hp}")
print(f"Fainted? {incineroar.is_fainted}")
```
