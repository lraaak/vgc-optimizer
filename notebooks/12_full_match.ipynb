```python
import sys
import pandas as pd
import json
sys.path.append('../src')
from battle_engine import BattlePokemon, BattleState
from ai_agent import SmartVGCAgent
from full_simulator import FullGameSimulator
from synergy_scorer import TeamSynergyScorer
```

```python
df_pokemon = pd.read_csv('../data/processed/clustered_pokemon_data.csv').set_index('name', drop=False)
df_moves = pd.read_csv('../data/raw/showdown_moves.csv').set_index('name', drop=False)
with open('../data/raw/type_chart.json', 'r') as f:
    type_chart = json.load(f)

def get_moves(*names):
    return [df_moves.loc[n].to_dict() for n in names]
```

```python
# Team A (Hyper Offense)
incineroar = BattlePokemon('incineroar', df_pokemon.loc['incineroar'], get_moves('Fake Out', 'Flare Blitz', 'Snarl'), 'Intimidate')
rillaboom = BattlePokemon('rillaboom', df_pokemon.loc['rillaboom'], get_moves('Wood Hammer', 'Grassy Glide', 'Protect'), 'Grassy Surge')
urshifu = BattlePokemon('urshifu-rapid-strike', df_pokemon.loc['urshifu-rapid-strike'], get_moves('Close Combat', 'Surging Strikes'), 'Unseen Fist')
tornadus = BattlePokemon('tornadus-incarnate', df_pokemon.loc['tornadus-incarnate'], get_moves('Bleakwind Storm', 'Tailwind'), 'Prankster')

# Team B (Bulky/Control)
flutter_mane = BattlePokemon('flutter-mane', df_pokemon.loc['flutter-mane'], get_moves('Dazzling Gleam', 'Icy Wind', 'Protect'), 'Protosynthesis')
chi_yu = BattlePokemon('chi-yu', df_pokemon.loc['chi-yu'], get_moves('Heat Wave', 'Dark Pulse'), 'Beads of Ruin')
amoonguss = BattlePokemon('amoonguss', df_pokemon.loc['amoonguss'], get_moves('Sludge Bomb', 'Pollen Puff'), 'Regenerator')
iron_hands = BattlePokemon('iron-hands', df_pokemon.loc['iron-hands'], get_moves('Drain Punch', 'Wild Charge', 'Fake Out'), 'Quark Drive')
```

```python
state = BattleState()
state.team_a_active = [urshifu, rillaboom]
state.team_a_bench = [incineroar, tornadus]

state.team_b_active = [flutter_mane, chi_yu]
state.team_b_bench = [amoonguss, iron_hands]

agent_a = SmartVGCAgent(None) 
agent_b = SmartVGCAgent(None)

sim = FullGameSimulator(type_chart)
```

```python
agent_a.engine = __import__('battle_engine').TurnEngine(state, type_chart)
agent_b.engine = __import__('battle_engine').TurnEngine(state, type_chart)

winner = sim.play_match(state, agent_a, agent_b)
```

```python
from monte_carlo import MonteCarloSimulator


team_a = [incineroar, rillaboom, urshifu, tornadus]
team_b = [flutter_mane, chi_yu, amoonguss, iron_hands]

print("Running 100 Monte Carlo Simulations. Please wait...")

# Initialize and Run!
mc_sim = MonteCarloSimulator(type_chart)
results = mc_sim.run_simulations(team_a, team_b, num_simulations=100)

print("=== MONTE CARLO RESULTS ===")
print(f"Team A Wins: {results['team_a_wins']}")
print(f"Team B Wins: {results['team_b_wins']}")
print(f"Team A Win Rate: {results['team_a_win_rate']}%")
```
