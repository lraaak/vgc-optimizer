```python
import sys
import pandas as pd
sys.path.append('../src')
from battle_engine import BattlePokemon, BattleState, TurnEngine
from synergy_scorer import TeamSynergyScorer
```

```python
df_pokemon = pd.read_csv('../data/processed/clustered_pokemon_data.csv').set_index('name')
df_moves = pd.read_csv('../data/raw/showdown_moves.csv').set_index('name', drop=False)

scorer = TeamSynergyScorer()
type_chart = scorer.type_chart

incineroar = BattlePokemon('incineroar', df_pokemon.loc['incineroar'], [], 'Intimidate')
flutter_mane = BattlePokemon('flutter-mane', df_pokemon.loc['flutter-mane'], [], 'Protosynthesis')
```

```python
state = BattleState()
state.team_a_active = [incineroar]
state.team_b_active = [flutter_mane]

engine = TurnEngine(state, type_chart)

flare_blitz = df_moves.loc['Flare Blitz'].to_dict()
moonblast = df_moves.loc['Moonblast'].to_dict()

actions = [
    {'user': incineroar, 'target': flutter_mane, 'move': flare_blitz},
    {'user': flutter_mane, 'target': incineroar, 'move': moonblast}
]
```

```python
print("--- EXECUTING TURN 1 ---")
logs = engine.execute_turn(actions)

for log in logs:
    print(log)
```
